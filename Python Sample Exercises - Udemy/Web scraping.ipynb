```python
# Now loop through all pages and extract titles of all books with 2-star ratings

url = 'https://books.toscrape.com/catalogue/page-{}.html'
base_url = url.format(2)

```

```python
twostarbooks = []

for p in range (51):
    res = requests.get(url.format(p))
    soup=bs4.BeautifulSoup(res.text,'lxml')
    books = soup.select('.product_pod')
    for each_book in books:
        if each_book.select('.star-rating.Two') != []:
            twostarbooks.append(each_book.select('a')[1]['title'])
        else:
            pass

        
print(twostarbooks)
```
